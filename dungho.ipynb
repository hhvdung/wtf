```
# Import the movie data for the project
mv = pd.read_csv('/content/drive/My Drive/FTMLE - Tonga/Data/movie_review.csv', encoding='utf-8', sep='\t')
```

```
mv.info
```

```
#Remove every special character on the review column
mv['review'] = mv['review'].map(lambda x: re.sub(r'\W+', ' ', x))
```

```
# Check to see if any error data
mv.isnull().sum()
```

```
# Convert the review column to numpy array also lowercase them
review = mv['review'].str.lower().to_numpy()
print(review)
```

# Accuracy score

```
_# Split data to train and test set
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import *

tf = TfidfVectorizer()
np.set_printoptions(precision=2)
tf_vec = tf.fit_transform(review)

X = mv['review']
y = mv['sentiment']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```
# Create a Logistic Regression model
from sklearn.pipeline import *
from sklearn.linear_model import *

clf = Pipeline([('vect', tf), ('clf', LogisticRegression(random_state =0))])
# Train the model with data
clf.fit(X_train, y_train)
```

```
# Accuracy of entire Data Frame
predictions = clf.predict(X)
accuracy_score(y, predictions)
```

```
# Add the prediction to the Data Frame
mv['Evaluation'] = clf.predict(X)
```
